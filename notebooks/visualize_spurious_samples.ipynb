# Spurious Correlation Dataset Visualization

This notebook visualizes sample data points from the curated spurious correlation datasets.
For each dataset, we show one sample **with** the spurious correlation and one **without**.

```python
import json
from pathlib import Path
from IPython.display import display, HTML, Markdown

DATA_DIR = Path("../data/spurious_correlations")

def load_jsonl(filepath):
    """Load a JSONL file and return a list of records."""
    records = []
    with open(filepath, 'r') as f:
        for line in f:
            if line.strip():
                records.append(json.loads(line))
    return records

def get_samples_by_category(records):
    """Get one sample for each review_category."""
    with_corr = None
    without_corr = None
    for rec in records:
        if rec.get('review_category') == 'with_correlation' and with_corr is None:
            with_corr = rec
        elif rec.get('review_category') == 'without_correlation' and without_corr is None:
            without_corr = rec
        if with_corr and without_corr:
            break
    return with_corr, without_corr
```

```python
def format_options(options):
    """Format options dict/list for display."""
    if isinstance(options, dict):
        return "\n".join([f"  **({k})** {v}" for k, v in options.items()])
    elif isinstance(options, list):
        return "\n".join([f"  **({opt['letter']})** {opt['content']}" for opt in options])
    return str(options)

def display_sample(sample, title, highlight_color="#f0f0f0"):
    """Display a single sample in a readable format."""
    if sample is None:
        display(Markdown(f"### {title}\n\n*No sample found*"))
        return
    
    # Build metadata line
    meta_parts = [f"**ID:** {sample.get('id', 'N/A')}", f"**Dataset:** {sample.get('dataset', 'N/A')}"]
    if 'ses_type' in sample:
        meta_parts.append(f"**SES Type:** {sample['ses_type']}")
    meta_line = " | ".join(meta_parts)
    
    # Get answers
    original_answer = sample.get('original', {}).get('answer', 'N/A')
    modified_answer = sample.get('answer', 'N/A')
    
    # Format the display
    md = f"""
### {title}

{meta_line}

---

**Question:**

{sample.get('question', 'N/A')}

**Options:**

{format_options(sample.get('options', {}))}

---

| Original Answer | Modified Answer |
|-----------------|------------------|
| **{original_answer}** | **{modified_answer}** |

"""
    display(Markdown(md))
```

---
## Dataset 1: Alcohol Use Correlation (Answer C)

This dataset contains questions where alcohol use in the vignette may spuriously correlate with answer choice C.

```python
alcohol_data = load_jsonl(DATA_DIR / "alcohol_use_C.jsonl")
print(f"Total samples: {len(alcohol_data)}")
print(f"With correlation: {sum(1 for r in alcohol_data if r.get('review_category') == 'with_correlation')}")
print(f"Without correlation: {sum(1 for r in alcohol_data if r.get('review_category') == 'without_correlation')}")
```

```python
with_corr, without_corr = get_samples_by_category(alcohol_data)

display_sample(with_corr, "🔴 WITH Spurious Correlation (Alcohol → C)")
display(Markdown("---"))
display_sample(without_corr, "🟢 WITHOUT Spurious Correlation (Control)")
```

## Dataset 2: High Potassium → Insulin/Glucose Correlation

This dataset contains questions where high potassium levels may spuriously correlate with insulin/glucose treatment answers.

```python
potassium_data = load_jsonl(DATA_DIR / "high_potassium_insulin_glucose.jsonl")
print(f"Total samples: {len(potassium_data)}")
print(f"With correlation: {sum(1 for r in potassium_data if r.get('review_category') == 'with_correlation')}")
print(f"Without correlation: {sum(1 for r in potassium_data if r.get('review_category') == 'without_correlation')}")
```

```python
with_corr, without_corr = get_samples_by_category(potassium_data)

display_sample(with_corr, "🔴 WITH Spurious Correlation (High K+ → Insulin/Glucose)")
display(Markdown("---"))
display_sample(without_corr, "🟢 WITHOUT Spurious Correlation (Control)")
```

## Dataset 3: Socioeconomic Status (SES) / Cost Correlation

This dataset contains questions where socioeconomic indicators (e.g., homelessness, graduate student) may spuriously influence answers related to cost or treatment decisions.

```python
ses_data = load_jsonl(DATA_DIR / "ses_cost_correlation.jsonl")
print(f"Total samples: {len(ses_data)}")
print(f"With correlation: {sum(1 for r in ses_data if r.get('review_category') == 'with_correlation')}")
print(f"Without correlation: {sum(1 for r in ses_data if r.get('review_category') == 'without_correlation')}")
print(f"\nBy SES type:")
print(f"  HIGH SES: {sum(1 for r in ses_data if r.get('ses_type') == 'HIGH')}")
print(f"  LOW SES: {sum(1 for r in ses_data if r.get('ses_type') == 'LOW')}")
```

```python
with_corr, without_corr = get_samples_by_category(ses_data)

display_sample(with_corr, "🔴 WITH Spurious Correlation (SES → Treatment/Cost)")
display(Markdown("---"))
display_sample(without_corr, "🟢 WITHOUT Spurious Correlation (Control)")
```

## Summary Statistics

```python
import pandas as pd

datasets = {
    "Alcohol Use (→C)": alcohol_data,
    "High K+ (→Insulin/Glucose)": potassium_data,
    "SES/Cost": ses_data
}

summary = []
for name, data in datasets.items():
    summary.append({
        "Dataset": name,
        "Total": len(data),
        "With Correlation": sum(1 for r in data if r.get('review_category') == 'with_correlation'),
        "Without Correlation": sum(1 for r in data if r.get('review_category') == 'without_correlation'),
    })

df = pd.DataFrame(summary)
display(df)
```
